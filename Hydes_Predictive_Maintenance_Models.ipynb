```python
err_19 = pd.read_csv("/Users/Jake/Google Drive/renewablesAIProject/Python_code/EC_19.csv", parse_dates=['datetime'], index_col=['datetime'])
err_19  = err_19.drop(['Unnamed: 0'], axis=1)

e1 = err_19.loc[err_19['inverter'] == 1]
e2 = err_19.loc[err_19['inverter'] == 2]
e3 = err_19.loc[err_19['inverter'] == 3]
e4 = err_19.loc[err_19['inverter'] == 4]
e5 = err_19.loc[err_19['inverter'] == 5]
e6 = err_19.loc[err_19['inverter'] == 6]

e1 = e1.resample('1H').agg({'inverter': np.max, 'Error_Code': np.max, 'Modules_affected': np.sum})
e2 = e2.resample('1H').agg({'inverter': np.max, 'Error_Code': np.max, 'Modules_affected': np.sum})
e3 = e3.resample('1H').agg({'inverter': np.max, 'Error_Code': np.max, 'Modules_affected': np.sum})
e4 = e4.resample('1H').agg({'inverter': np.max, 'Error_Code': np.max, 'Modules_affected': np.sum})
e5 = e5.resample('1H').agg({'inverter': np.max, 'Error_Code': np.max, 'Modules_affected': np.sum})
e6 = e6.resample('1H').agg({'inverter': np.max, 'Error_Code': np.max, 'Modules_affected': np.sum})
es = [e1, e2, e3, e4, e5, e6]

err_19 = pd.concat(es)
err_19 = err_19[np.isfinite(err_19['inverter'])]


scada_cleaned = scada_cleaned.reset_index()
err_19 = err_19.reset_index()
scada_err_19 = pd.merge(scada_cleaned, err_19, how='outer', on=['datetime', 'inverter'])

scada_err_19['Error_Code'] = scada_err_19['Error_Code'].fillna(0)
scada_err_19['Modules_affected'] = scada_err_19['Modules_affected'].fillna(0)
scada_err_19['Error_Code'] = scada_err_19['Error_Code'].shift(-1)
scada_err_19 = scada_err_19.rename(columns={'Error_Code': 'Err_Code_nxt_hour'})
```

```python
from sklearn.cross_validation import train_test_split
import random
scada_err_19 = scada_err_19.dropna()
scada_err_19.drop('datetime', axis=1, inplace=True)
X = scada_err_19.drop('Err_Code_nxt_hour', axis=1) 
y = scada_err_19['Err_Code_nxt_hour']
random.seed(40)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, stratify=y)
```

## Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix 
import random 
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=20, min_samples_leaf=5)

clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100,
                               max_depth=20, min_samples_leaf=5)
clf_gini.fit(X_train, y_train) 
clf_entropy.fit(X_train, y_train) 
y_pred = clf_gini.predict(X_test) 
print(confusion_matrix(y_test,y_pred))  
print(classification_report(y_test,y_pred))
y_pred = clf_entropy.predict(X_test) 
print(confusion_matrix(y_test,y_pred))  
print(classification_report(y_test,y_pred))
```

## Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(X_train, y_train)
predictions = rf.predict(X_test)
predictions = pd.DataFrame(predictions)
predictions[predictions.columns[0]]= predictions[predictions.columns[0]].round()
print(confusion_matrix(y_test, predictions))  
print(classification_report(y_test, predictions))
```

## KNN Classifier Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)
print(confusion_matrix(y_test, predictions))  
print(classification_report(y_test, predictions))
```

## XGBoost Model

```python
import xgboost as xgb
from xgboost import XGBClassifier
from xgboost import plot_importance
import matplotlib.pyplot as plt

XGBmodel = XGBClassifier()
XGBmodel.fit(X_train,y_train)
predictions = XGBmodel.predict(X_test) 
print(confusion_matrix(y_test, predictions))  
print(classification_report(y_test, predictions))

xgb.plot_importance(XGBmodel)
```

## Optimised XGBoost

```python
from sklearn.cross_validation import *
from sklearn.grid_search import GridSearchCV
cv_params = {'max_depth': [3,5,7], 'min_child_weight': [1,3,5]}
ind_params = {'learning_rate': 0.1, 'n_estimators': 1000, 'seed':0, 'subsample': 0.8, 'colsample_bytree': 0.8, 
             'objective': 'binary:logistic'}

optimized_GBM = GridSearchCV(XGBClassifier(**ind_params), 
                            cv_params, 
                             scoring = 'recall', cv = 5, n_jobs = -1) 

optimized_GBM .fit(X_train,y_train)
```
